### Focus on total enrollment and grades pk through grade 12

### Creating heatmap

```python
# Creating correlation table between variables
df_sub.corr()
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr_1 = sns.heatmap(df_sub.corr(), annot = True, ax = ax)
```

#### Insight from heatmap: 
There are not any particularly strong correlations between these variables - Total Enrollment is positively correlated with all grade levels (PK to 12)suggesting a natural progression of student enrollment from one grade to the next. The next correlation Grade 9, Grade 10, Grade 11, and Grade 12 show strong positive correlations with each other, indicating consistent enrollment patterns across these grade levels.

### Focus on total enrollment, race, SWD, ELL, and poverty

```python
# Creating data subset with variables to focus on
df_sub2 = df_clean[['% Female', '% Male', '% Asian', '% Black', '% Hispanic', '% Multiple Race Categories Not Represented','% White', '% Students with Disabilities', '% English Language Learners', '% Poverty']]
```

```python
df_sub2
```

```python
# Creating correlation table between variables
df_sub2.corr()
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr_1 = sns.heatmap(df_sub2.corr(), annot = True, ax = ax)
```

#### Insight from heatmap: 
- Strong negative correlation between % Female and % Male (-0.999998), indicating that as the percentage of females increases, the percentage of males decreases and vice versa.
- Negative correlations between % Asian and % Black (-0.448174), % Asian and % Hispanic (-0.314513), and % Black and % Hispanic (-0.459035), indicating that higher representation of one ethnicity tends to be associated with lower representation of others.
- Strong negative correlation between % White and % Students with Disabilities and English Language Learner, suggesting lower representation of students with disabilities in predominantly white populations.
