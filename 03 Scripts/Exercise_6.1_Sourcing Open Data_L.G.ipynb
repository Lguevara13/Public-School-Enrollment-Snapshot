#### There are no missing values in the dataset to remove. It has been observed that the variable 'Economic Need Index' contains values that says No date instead of NaN. No changes are needed. 

```python
# Looking for mixed data types
for col in df.columns.tolist():
    weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
    if len (df[weird]) > 0:
        print (col)
```

#### There is no mixed data
